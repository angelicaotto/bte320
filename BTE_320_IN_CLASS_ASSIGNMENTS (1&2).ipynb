```python
Number1 = float(input("Enter Number 1:"))
Number2 = float(input("Enter Number 2:"))
Operator = input("Enter operator (+, -, *, /, %, **, //):")

if Operator == "+":
    print(Number1 + Number2)
elif Operator == "-":
    print(Number1 - Number2)
elif Operator == "*":
    print(Number1 * Number2)
elif Operator == "/":
    print(Number1 / Number2)
elif Operator == "%":
    print(Number1 % Number2)
elif Operator == "**":
    print(Number1 ** Number2)
elif Operator == "//":
    print(Number1 // Number2)
else:
    print("Invalid Operator")
```
